## YOLO V11 test

```python
import os
import time
import cv2
from pathlib import Path
from ultralytics import YOLO  

def Yolo_predict(model, path, output_folder, SAVE_IMAGES=False):

    output_folder.mkdir(parents=True, exist_ok=True)
    total_time = 0
    img_count = 0

    for file_name in sorted(path.glob("*")):

        img = cv2.imread(str(file_name))
        if img is None:
            print(f"[!] No se pudo cargar: {file_name}")
            continue

        # Tiempo por imagen
        start_time = time.time()
        results = model(img, conf=0.5)
        elapsed = time.time() - start_time
        
        total_time += elapsed
        img_count += 1
        print(f"Imagen: {file_name.name}  |  Tiempo: {elapsed:.4f} s")

        # Mostrar solo personas (clase 0 en COCO)
        for result in results:
            for box in result.boxes:
                cls = int(box.cls.item())
                if cls == 0:  # 0 es 'person' en YOLO COCO
                    x1, y1, x2, y2 = map(int, box.xyxy[0])
                    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Guardar imagen procesada
        if SAVE_IMAGES:
            output_img_path = output_folder / file_name.name
            cv2.imwrite(str(output_img_path), img)

    print("\n-------------------------------")
    print(f"Total imgenes: {img_count}")
    print(f"Tiempo total: {total_time:.4f} s")
    print(f"Tiempo promedio: {total_time/img_count:.4f} s")
    print("-------------------------------\n")


def main():
    model = YOLO("yolo11s.pt")
    path = Path("./Dataset/test/")
    output = Path("./Results/YOLOV11S")

    Yolo_predict(model, path, output, SAVE_IMAGES=True)

main()
```

## POSE DETECTION

```python
import os
import cv2
import numpy as np
from pathlib import Path
from ultralytics import YOLO
import time

def run_pose_inference(model, input_folder, output_folder, log_file):

    os.makedirs(output_folder, exist_ok=True)
    log = open(log_file, "w")

    total_start = time.time()

    # Iterar sobre las imgenes del dataset
    for img_path in sorted(Path(input_folder).glob("*")):
        start = time.time()

        # Cargar imagen original
        img = cv2.imread(str(img_path))
        
        img_padded = resize_keep_aspect(img)

        results = model.predict(
            source=img_padded,
            device="AUTO",       # Higher confidence threshold = fewer detections
            verbose=False,
            iou=0.7,        # NMS threshold
            half=True,      # FP16 inference (if supported)
            agnostic_nms=True  # Faster NMS
            
        )
        
        # Dibujar los resultados (YOLO ya trae mtodo de visualizacin)
        annotated_img = results[0].plot()

        # Guardar imagen de salida
        out_path = Path(output_folder) / img_path.name
        cv2.imwrite(str(out_path), annotated_img)

        # Obtener detecciones y keypoints
        boxes = results[0].boxes
        keypoints = results[0].keypoints
        
        if boxes is None or len(boxes) == 0:
            log.write(f"{img_path.name}: No detections\n")
            continue
        
        # Loguear informacin
        log.write(f"\n=== {img_path.name} ===\n")
        for i, (box, kpts_xy, kpts_conf) in enumerate(
            zip(boxes, keypoints.xy, keypoints.conf)
        ):
            x1, y1, x2, y2 = box.xyxy[0].tolist()
            conf = box.conf[0].item()
            log.write(f"Person {i+1}: box=({x1:.1f}, {y1:.1f}, {x2:.1f}, {y2:.1f}), conf={conf:.3f}\n")

            for j, (xy, c) in enumerate(zip(kpts_xy, kpts_conf)):
                kx, ky = xy
                log.write(f"   Keypoint {j:02d}: ({kx:.1f}, {ky:.1f}), conf={c:.3f}\n")

        end = time.time()
        print(f"{img_path.name} processed in {(end-start)*1000:.2f} ms")

    total_end = time.time()
    total_time = total_end - total_start
    print(f"\n Total processing time: {total_time:.2f} seconds")
    log.write(f"\nTotal processing time: {total_time:.2f} seconds\n")
    log.close()

def resize_keep_aspect(img, target_size=640):
    h, w = img.shape[:2]
    scale = target_size / max(h, w)
    new_w, new_h = int(w * scale), int(h * scale)
    
    # Use INTER_NEAREST for faster resizing (slight quality loss)
    resized = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_NEAREST)
    
    pad_x, pad_y = target_size - new_w, target_size - new_h
    padded = cv2.copyMakeBorder(resized, 0, pad_y, 0, pad_x,cv2.BORDER_CONSTANT, value=(114, 114, 114))
    return padded


def main():
    #model = YOLO("yolo11n-pose.pt")  # Modelo YOLO Pose
    ov_model = YOLO('yolo11n-pose_int8_openvino_model')  # OpenVino model 

    dummy_image = np.zeros((320, 320, 3), dtype=np.uint8)

    print("Calentando el modelo...")
    _ = ov_model.predict(source=dummy_image, imgsz=320)
    print("Calentamiento completado.")
    
    input_folder = "./Dataset/Tests/"   # Carpeta con tus imgenes
    output_folder = "./Results/NormalModel/Test8/Images/"
    log_file = "./Results/NormalModel/Test8/logs/logs.txt"
    run_pose_inference(ov_model, input_folder, output_folder, log_file)
```

```python
import numpy as np
import cv2
import time
from ultralytics import YOLO

def run_pose_inference(model, input_frame):
    # Remover streaming=True para obtener resultado directo
    results = model.predict(
        source=input_frame,
        device="cpu",        # Cambiado de AUTO
        verbose=False,
        iou=0.7,
        half=False,          # Desactivado para OpenVINO INT8
        agnostic_nms=True,
        conf=0.25            # Aadido threshold de confianza
    )
    
    # results es una lista, tomar el primer elemento
    result = results[0]
    annotated_img = result.plot()
    boxes = result.boxes
    keypoints = result.keypoints
    
    # Verificacin corregida
    if boxes is None or len(boxes) == 0:
        return annotated_img, None, None
    
    return annotated_img, boxes, keypoints

def resize_keep_aspect(img, target_size=640):
    h, w = img.shape[:2]
    scale = target_size / max(h, w)
    new_w, new_h = int(w * scale), int(h * scale)
    
    resized = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_NEAREST)
    
    pad_x, pad_y = target_size - new_w, target_size - new_h
    padded = cv2.copyMakeBorder(
        resized, 0, pad_y, 0, pad_x, 
        cv2.BORDER_CONSTANT, 
        value=(114, 114, 114)
    )
    return padded

def show_camera_with_model(camera_index=0, model=None):

    # Saltar frames para mayor FPS
    frame_skip = 2
    frame_count = 0      
    
    cap = cv2.VideoCapture(camera_index)
    
    if not cap.isOpened():
        print("No se pudo acceder a la cmara.")
        return
    
    # Configurar resolucin nativa para mejor performance
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
    cap.set(cv2.CAP_PROP_FPS, 30)  # Opcional: fijar FPS
    
    prev_frame_time = 0
    font = cv2.FONT_HERSHEY_SIMPLEX  # Mover fuera del loop
    
    print("Presiona 'q' para salir...")
    
    while cap.isOpened():
        ret, frame = cap.read()
        frame_count += 1

        if not ret:
            print("Error al leer frame")
            break
        
        if frame_count % frame_skip != 0:
            continue  # Procesar solo cada N frames
                
        # Redimensionar manteniendo aspecto
        resized_frame = resize_keep_aspect(frame, target_size=640)
        
        # Inferencia
        output_frame, boxes, keypoints = run_pose_inference(model, resized_frame)
        
        # Calcular FPS
        new_frame_time = time.time()
        fps = 1 / (new_frame_time - prev_frame_time) if prev_frame_time > 0 else 0
        prev_frame_time = new_frame_time
        
        # Mostrar FPS
        cv2.putText(
            output_frame, 
            f"FPS: {int(fps)}", 
            (7, 40),  # Ajustado para mejor visibilidad
            font, 
            1,  # Tamao ms pequeo
            (0, 255, 0), 
            2, 
            cv2.LINE_AA
        )
        
        # Mostrar cantidad de detecciones
        num_detections = len(boxes) if boxes is not None else 0
        cv2.putText(
            output_frame, 
            f"Personas: {num_detections}", 
            (7, 70), 
            font, 
            1, 
            (0, 255, 0), 
            2, 
            cv2.LINE_AA
        )
        
        cv2.imshow('YOLO11 Pose Detection', output_frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()
```

```python
ov_model = YOLO('yolo11n-pose_int8_openvino_model')
    
# Calentamiento con tamao consistente
print("Calentando el modelo...")
dummy_image = np.zeros((640, 640, 3), dtype=np.uint8)
_ = ov_model.predict(source=dummy_image, verbose=False)
print("Calentamiento completado.\n")
    
show_camera_with_model(0, ov_model)
```

```python
import os
import cv2
import numpy as np
from pathlib import Path
from ultralytics import YOLO
import time

def run_pose_inference(model, input_video, output_folder, log_file):
    """
    Extrae keypoints de un video y los guarda en formato STGCN.
    Formato: frame_id, person_id, x1, y1, conf1, x2, y2, conf2, ...
    """
    os.makedirs(output_folder, exist_ok=True)
    
    cap = cv2.VideoCapture(str(input_video))
    if not cap.isOpened():
        print(f" Error: No se puede abrir el video {input_video}")
        return
    
    # Obtener informacin del video
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    print(f" Video: {width}x{height} @ {fps:.2f} FPS, {total_frames} frames")
    
    frame_count = 0
    txt_count = 0
    detected_persons = set()
    
    with open(log_file, "w") as log:
        while True:
            ret, frame = cap.read()
            
            if not ret:
                break
            
            # Resize manteniendo aspecto
            resized_frame = resize_keep_aspect(frame)
            
            # Inferencia YOLO Pose
            results = model.predict(
                source=resized_frame,
                device="AUTO",
                verbose=False,
                iou=0.7,
                half=True,
                agnostic_nms=True,
                conf=0.3  # Umbral de confianza para detecciones
            )
            
            boxes = results[0].boxes
            keypoints = results[0].keypoints
            
            # Si no hay detecciones, no escribir nada (o frame vaco)
            if boxes is None or len(boxes) == 0:
                frame_count += 1
                continue
            
            # Procesar cada persona detectada
            for person_id, (box, kpts_xy, kpts_conf) in enumerate(zip(boxes, keypoints.xy, keypoints.conf)):
                detected_persons.add(person_id)
                
                # Normalizar coordenadas al rango [0, 1]
                # YOLO devuelve coords en el espacio de la imagen resized (640x640)
                normalized_kpts = []
                
                for (x, y), conf in zip(kpts_xy, kpts_conf):
                    # Normalizar x, y al rango [0, 1]
                    x_norm = x / 640.0
                    y_norm = y / 640.0
                    
                    # Clamp entre 0 y 1 por seguridad
                    x_norm = max(0.0, min(1.0, x_norm))
                    y_norm = max(0.0, min(1.0, y_norm))
                    
                    # Confianza ya est en [0, 1]
                    conf_val = float(conf)
                    
                    normalized_kpts.extend([x_norm, y_norm, conf_val])
                
                # Escribir: frame_id, person_id, x1, y1, conf1, x2, y2, conf2, ...
                log.write(f"{frame_count:03d},{person_id},")
                log.write(",".join(f"{v:.3f}" for v in normalized_kpts))
                log.write("\n")
            
            frame_count += 1
            
            # Progreso cada 100 frames
            if frame_count % 20 == 0:
                txt_count += 1
                
                nombre, extension = os.path.splitext(archivo)
                nuevo_nombre = f"{nombre}_1{extension}"
                
                print(nuevo_nombre)  # logs_1.txt 
                
                print(f" Procesados {frame_count}/{total_frames} frames...")
    
    cap.release()
    
    print(f"\n Procesamiento completo:")
    print(f"    Frames totales: {frame_count}")
    print(f"    Personas nicas detectadas: {len(detected_persons)}")
    print(f"    Log guardado en: {log_file}")


def resize_keep_aspect(img, target_size=640):
    """
    Redimensiona manteniendo aspecto y aade padding.
    """
    h, w = img.shape[:2]
    scale = target_size / max(h, w)
    new_w, new_h = int(w * scale), int(h * scale)
    
    # INTER_LINEAR es mejor que INTER_NEAREST para pose
    resized = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    
    # Padding para llegar a target_size x target_size
    pad_x, pad_y = target_size - new_w, target_size - new_h
    padded = cv2.copyMakeBorder(
        resized, 0, pad_y, 0, pad_x,
        cv2.BORDER_CONSTANT, value=(114, 114, 114)
    )
    
    return padded


def main():
    print(" Iniciando extraccin de keypoints para STGCN\n")
    
    # Cargar modelo OpenVINO optimizado
    ov_model = YOLO('yolo11n-pose_int8_openvino_model')
    
    # Calentar el modelo
    print(" Calentando modelo...")
    dummy_image = np.zeros((640, 640, 3), dtype=np.uint8)
    _ = ov_model.predict(source=dummy_image, imgsz=640, verbose=False)
    print(" Modelo listo\n")
    
    # Rutas
    input_video = './Models/Action/mmaction2/demo/demo_skeleton.mp4'
    
    output_folder = "./Results/RealTime/Test3/logs"
    log_file = "./Results/RealTime/Test3/logs/logs.txt"

    # Verificar que el video existe
    if not os.path.exists(input_video):
        print(f" Error: Video no encontrado en {input_video}")
        return
    
    # Procesar video
    start_time = time.time()
    run_pose_inference(ov_model, input_video, output_folder, log_file)
    elapsed = time.time() - start_time
    
    print(f"\n Tiempo total: {elapsed:.2f} segundos")
    print(f" FPS promedio: {elapsed/100:.2f}" if elapsed > 0 else "N/A")


if __name__ == '__main__':
    main()
```

## ACTION 2S/AGCN

```python
import numpy as np
from mmaction.apis import init_recognizer, inference_recognizer
import torch

import numpy as np

def load_keypoints_from_txt_multi_person(path, num_joints=17):
    """
    Carga keypoints multi-persona desde un archivo txt en formato YOLO:
    frame_id, person_id, x1, y1, conf1, x2, y2, conf2, ...
    
    Devuelve un array (M, T, V, C) -> (n_personas, n_frames, n_joints, coordenadas)
    listo para usar con ST-GCN (MMAction2).
    """
    frames_dict = {}  # frame_id -> person_id -> keypoints array (V,3)
    
    with open(path, 'r') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
                
            parts = line.split(',')
            expected_values = 2 + (num_joints * 3)
            if len(parts) < expected_values:
                print(f" Lnea ignorada (valores insuficientes): {len(parts)} valores, esperados {expected_values}")
                continue
            
            try:
                frame_id = int(parts[0])
                person_id = int(parts[1])
                data = list(map(float, parts[2:2 + num_joints * 3]))
                kpts = np.array(data).reshape(num_joints, 3)
                
                if frame_id not in frames_dict:
                    frames_dict[frame_id] = {}
                frames_dict[frame_id][person_id] = kpts
                
            except (ValueError, IndexError) as e:
                print(f" Error procesando lnea: {e}")
                continue

    if not frames_dict:
        raise ValueError(" No se pudieron cargar keypoints del archivo")
    
    # Dimensiones globales
    T = len(frames_dict)  # nmero de frames
    M = max(pid for persons in frames_dict.values() for pid in persons) + 1  # nmero de personas
    
    print(f" Datos cargados: {T} frames, {M} personas, {num_joints} joints")
    
    # Crear array (T, M, V, C)
    frames = np.zeros((T, M, num_joints, 3), dtype=np.float32)
    
    for t, (frame_id, persons) in enumerate(sorted(frames_dict.items())):
        for pid, kpts in persons.items():
            frames[t, pid, :, :] = kpts

    # Reordenar a formato esperado por MMAction2 (M, T, V, C)
    frames = np.transpose(frames, (1, 0, 2, 3))
    
    print(f" Shape final (M, T, V, C): {frames.shape}")
    return frames



def main():
    #config_file = './Models/Action/mmaction2/configs/skeleton/stgcn/stgcn_8xb16-bone-motion-u100-80e_ntu60-xsub-keypoint-2d.py'
    #checkpoint_file = './Models/Action/mmaction2/checkpoints/stgcn_80e_ntu60_xsub_keypoint-e7bb9653.pth'

    config_file = './Models/Action/mmaction2/configs/skeleton/2s-agcn/2s-agcn_8xb16-joint-u100-80e_ntu60-xsub-keypoint-2d.py'
    # download the checkpoint from model zoo and put it in `checkpoints/`
    checkpoint_file = './Models/Action/mmaction2/checkpoints/2s-agcn_8xb16-joint-u100-80e_ntu60-xsub-keypoint-2d_20221222-4c0ed77e.pth'
    
    
    print(" Inicializando modelo STGCN...")
    model = init_recognizer(config_file, checkpoint_file, device='cpu')
    
    print(" Cargando keypoints...")
    keypoints = load_keypoints_from_txt_multi_person(
        './Results/RealTime/Test3/logs/logs.txt'
    )

    results = {
    'keypoint': keypoints,  # (M, T, V, C)
    'total_frames': keypoints.shape[1]  # T
    }

    
    print(" Realizando inferencia...")
    result = inference_recognizer(model, results)
    
    print("\n" + "="*50)
    print(" RESULTADO DE ACCIN")
    print("="*50)
    
    # Extraer prediccin principal
    pred_scores = result.pred_score.tolist()
    pred_label = result.pred_label.item()
    
    # Mostrar top-5 predicciones
    top5_indices = np.argsort(pred_scores)[::-1][:5]
    
    print(f"\n Accin predicha: Clase {pred_label} (score: {pred_scores[pred_label]:.4f})")
    print("\n Top 5 acciones:")
    for i, idx in enumerate(top5_indices, 1):
        print(f"  {i}. Clase {idx}: {pred_scores[idx]:.4f}") 
```
